## Water usage revisited
In this activity, we will use a tree map to visualize the water usage.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import squarify
```

Use pandas to read the data located in the subfolder data.

```python
mydata = pd.read_csv("data/water_usage.csv")
```

```python
mydata
```

Use a tree map to visualize the water usage. Show the percentages for each tile and add a title.

```python
# Create figure
plt.figure(dpi=200)
# Create tree map
labels = mydata['Usage'] + ' (' + mydata['Percentage'].astype('str') + '%)'
squarify.plot(sizes=mydata['Percentage'], label=labels, color=sns.light_palette('green', mydata.shape[0]))
plt.axis('off')
# Add title
plt.title('Water usage')
# Show plot
plt.show()
```
