## APIs simples

### Brasil API
Brasil API é uma iniciativa para consolidação de dados públicos.<br>
<strong>O que fazer</strong>: acesso e download<br>
<strong>Documentação</strong>: https://brasilapi.com.br/docs#tag/Brasil-API<br>
<strong>Objetivo</strong>: Qual o meu endereço? Meu CEP é 04026-090?<br>

```python
import requests
import pandas as pd
```

```python

cep = '04026-090'
response = requests.get(f"https://brasilapi.com.br/api/cep/v1/{cep}")
endereco = response.json()
endereco
print(f"O meu é endereço com o cep {cep} é",endereco["neighborhood"],",",endereco["city"])
```

### Mercado Livre
Vamos utilizar a API do MercadoLivre para consultar anúncios<br>
<strong>O que fazer</strong>: acesso, download, tratamento e cálculo<br>
<strong>Documentação</strong>: https://developers.mercadolivre.com.br/pt_br/api-docs-pt-br<br>
<strong>Objetivo</strong>: Qual o preço médio dos anúncios de iphones em Minas Gerais<br>
<strong>Desafio</strong>: Resolve o exercício através de uma função para permitir realizar a mesma consulta com outros itens ou estados<br>
<strong>Dica</strong>: Carregue seus dados para um dataframe para facilitar a manipulação dos dados

```python
response = requests.get("https://api.mercadolibre.com/sites/MLB/search?q=Iphone&state=BR-MG&category=MLB1055")
data = response.json()
data
```

```python
dt = pd.DataFrame(data['results'])
preco = round(dt['price'].mean(),2)
print(f'O valor médio de Iphone em Minas Gerais é {preco}')
```
